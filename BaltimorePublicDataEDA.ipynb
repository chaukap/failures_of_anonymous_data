```python
import pandas as pd
import numpy as np
import math
import plotly.express as px
```

```python
overdoses = pd.read_csv("Baltimore_City_Fire_Department_Clinician-Administered_Naloxone.csv", index_col="ObjectId")
overdoses.head()
```

```python
overdoses = overdoses[overdoses.BCFD_Clinician_Naloxone_Adminis.map(
    lambda y: not math.isnan(y)
)]
```

```python
df = overdoses.groupby("Incident_Date").count()
print(df.tail())
fig = px.line(df, x=df.index, y="Incident_Postal_Code", title='Number of postal codes with overdoses')
fig.show()
```
