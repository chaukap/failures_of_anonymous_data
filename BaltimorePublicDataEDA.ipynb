```python
import pandas as pd
import numpy as np
import math
import plotly.express as px
```

```python
overdoses = pd.read_csv("Baltimore_City_Fire_Department_Clinician-Administered_Naloxone.csv", index_col="ObjectId")
arrests = pd.read_csv("Arrests.csv")
arrests["Date"] = arrests.ArrestDateTime.map(lambda q: q[0:10])
substance_arrest = arrests[arrests.ChargeDescription.map(lambda t: "CDS" in t or "Poss" in t)]
leave_behind = pd.read_csv("Baltimore_City_Fire_Department_Leave-Behind_Naloxone_Distribution.csv")
calls = pd.read_csv("911_Calls_For_Service_2020.csv")
```

```python
overdoses = overdoses[overdoses.BCFD_Clinician_Naloxone_Adminis.map(
    lambda y: not math.isnan(y)
)]
leave_behind = leave_behind[leave_behind.Number_of_Leave_Behind_Naloxone.map(
    lambda y: not math.isnan(y)
)]
```

```python
leave_behind["Date"] = leave_behind.Incident_Date.map(lambda r: r[0:10])
overdoses["Date"] = overdoses.Incident_Date.map(lambda r: r[0:10])
calls["Date"] = calls.calldatetime.map(lambda y: y[0:10])
leave_behind.rename(columns={
    "Incident_Postal_Code": "zipcode"
}, inplace=True)
overdoses.rename(columns={
    "Incident_Postal_Code": "zipcode"
}, inplace=True)
```

```python
leave_behind.merge(calls, on=["Date", "zipcode"])[["Date", "zipcode", "location", "description", "priority", "Number_of_Leave_Behind_Naloxone"]]
```

```python
overdoses.merge(calls, on=["Date", "zipcode"])[["Date", "zipcode", "location", "description", "priority", "BCFD_Clinician_Naloxone_Adminis"]]
```
